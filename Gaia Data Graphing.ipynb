```python
dup_entry_num = 0
dup_entry_num_list = []

for entry in ID_GAIA:
    temp_num = 0
    for pot_dup in ID_GAIA:
        if entry == pot_dup:
            if temp_num != dup_entry_num:   
                dup_entry_num_list.append((dup_entry_num, temp_num))
                print(dup_entry_num, temp_num)
        temp_num += 1
    dup_entry_num += 1

print(dup_entry_num_list)
```

```python
entries_to_remove_gaia = []

for entry in dup_entry_num_list:
    x, y = entry
    if DIST[x] > DIST[y]:
        print("Keep " + str(x) + " and discard " + str(y))
        entries_to_remove.append(y)
    elif DIST[y] > DIST[x]:
        print("Keep " + str(y) + " and discard " + str(x))
    else:  #The two are equal
        print(str(x) + " and " + str(y) + " have equal distance.")

        
print(entries_to_remove_gaia)

    
```

```python
entries_to_remove_gaia = sorted(entries_to_remove_gaia)
```

```python
def remove_entries(LIST, entriesToRemove):
    temp_num = 0
    temp_list = []
    for entry in LIST:
        if temp_num not in entriesToRemove:
            temp_list.append(entry)
        temp_num += 1
    return(temp_list)
    
```

```python
ID_GAIA = remove_entries(ID_GAIA, entries_to_remove_gaia)
```

```python
dup_entry_num_list_maps = []
temp_num = 0
for entry in ID_MAPS:
    temp_num_2 = 0
    for dup_entry_check in ID_MAPS:
        if entry == dup_entry_check:
            if temp_num != temp_num_2:
                print(temp_num, temp_num_2, entry)
                dup_entry_num_list_maps.append((temp_num, temp_num_2))
        temp_num_2 += 1
    temp_num += 1
```

```python
entries_to_remove_maps = []

for entry in dup_entry_num_list_maps:
    x, y = entry
    if SNR[x] > SNR[y]:
        print("Keep " + str(x) + " and discard " + str(y))
        entries_to_remove.append(y)
    elif SNR[y] > SNR[x]:
        print("Keep " + str(y) + " and discard " + str(x))
    else:  #The two are equal
        print(str(x) + " and " + str(y) + " have equal SNR.")

entries_to_remove_maps.append(1474) 
entries_to_remove_maps.append(1578) 
entries_to_remove_maps.append(1753)
#entries that had no gaia counterpart
entries_to_remove_maps.append(1914)
entries_to_remove_maps.append(4114)
entries_to_remove_maps = sorted(entries_to_remove_maps)
print(entries_to_remove_maps)
```
